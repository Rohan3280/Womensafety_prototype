```python
def merge_dicts(dict_list):
    
    merged_dict = {}

    # Iterate over each dictionary in the list
    for d in dict_list:
        if isinstance(d, dict):  # Ensure the item is a dictionary
            merged_dict.update(d)

    return merged_dict

pred = merge_dicts(pred12)

print(pred)
```

```python
def merge_dicts_preserve_first(dict_list):
    merged_dict = {}
    for d in dict_list:
        for key, value in d.items():
            if key not in merged_dict:
                merged_dict[key] = value
    return merged_dict

pred2 = merge_dicts_preserve_first(pred13)
```
